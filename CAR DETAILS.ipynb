```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.Figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True,color='green')
plt.title('Hisplot For Year')
plt.xlabel('year')
plt.ylabel('count')
plt.show
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='seller_type', data=df, palette='viridis')
plt.title('selling_type count plot')
plt.xlabel('seller_type')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['year'].dropna(), bins=30, kde=True, color='blue')
plt.title('Year Distribution')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
cars=df['owner'].value_counts()
#piechar
plt.figure(figsize=(8,6))
plt.pie(cars, labels=['First Owner', 'Second Owner', 'Third Owner', 'Fourth & Above Owner', 'Test Drive Cars'],
        autopct='%1.1f%%', colors=['lightcoral', 'red', 'blue', 'royalblue'], startangle=180, explode=(0.05, 0.05,0.05,0.05,0.05))
plt.title('Survival Rate on Titanic')
plt.show()
```

```python
cars = df ['owner'].value_counts()
cars
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='seller_type', y='selling_price', data=df, palette='viridis')
plt.title('Fare Distribution by passenger class')
plt.xlabel('class')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='year', y='km_driven', hue='owner', style='seller_type', data= df, palette='rainbow')
plt.title('year vs km_driven with owner and seller_type')
plt.xlabel('year')
plt.ylabel('km_driven')
plt.show()
```

```python
num=df[["year", "selling_price", "km_driven"]].corr()
num
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(num, annot=True, cmap="rainbow", center=0)
plt.title("Corelation Heatmap of Car Dataset")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='fuel',y='year',hue='seller_type', data=df, split=True, palette='Set2')
plt.title("seller_type Distribution by fuel type")
plt.xlabel("fuel")
plt.ylabel("year")
plt.legend(title="survived")
plt.show()
```

```python
sns.kdeplot(data=df["year"].dropna(),color="skyblue",shade=True)
plt.title("year Distribution (kernel Density Estimate)")
plt.xlabel("year")
plt.ylabel=("Density")
plt.show()
```

```python
g=sns.FacetGrid(df,col='seller_type', row='owner',margin_titles=True)
g.map(sns.histplot, 'year', bins=20, color='teal')
g.fig.suptitle('age Distribution by class and survival', y=1.03)
plt.show()
```

```python
sns.pairplot(df[['year', 'selling_price', 'km_driven', 'transmission']],
            hue='transmission', palette='husl',
            plot_kws={'alpha':0.6},diag_kind='kde')
plt.suptitle('pair plot of titanic variables',y=1.02)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.swarmplot(x='fuel', y='km_driven', hue='transmission',data=df, palette='Set1', dodge=True)
plt.title('transmission distribution by fuel and km_driven(swarmplot)')
plt.xlabel('fuel')
plt.ylabel('km_driven')
plt.legend(title='transmission')
plt.show()
```
